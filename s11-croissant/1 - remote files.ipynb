```python
import os
import digitalhub as dh

p_name = f"digitalhub-tutorial-{os.getpid()}"
project = dh.get_or_create_project(p_name)
```

Create the dataitem with the path to the remote file

```python
path = "https://www.kaggle.com/datasets/uciml/adult-census-income/croissant/download"
dataitem = project.new_dataitem("kaggle-croissant", "croissant", path=path)
```

Transform the dataitem to a croissant dataset

```python
dataset = dataitem.as_dataset(overwrite=True)
```

Extract samples from the croissant dataset

```python
records = dataset.records(record_set="adult.csv")

for i, record in zip(range(5), records):
    print(record)
```
