```python
import os
import digitalhub as dh

p_name = f"digitalhub-tutorial-{os.getpid()}"
project = dh.get_or_create_project(p_name)
```

Create the dataitem with the path to the local file and log it to the project

```python
path = "data/croissant/metadata.json"
dataitem = project.log_croissant("local-croissant", drop_existing=True, source=path)
```

Transform the dataitem to a croissant dataset

```python
dataset = dataitem.as_dataset(overwrite=True)
```

Extract samples from the croissant dataset

```python
records = dataset.records(record_set="persons")

for i, record in zip(range(5), records):
    print(record)
```
